```python
import os
import sys
project_root = os.getcwd()
sys.path.append(project_root + '/modules')
# Service Pipeline의 구조를 이해하기 쉽게 하기 위해 작성한 ipynb 파일입니다
from Pipeline import pipe_service
pipe = pipe_service(verbose = True) # verbose = True로 설정할 경우, retrieval된 결과를 출력합니다.
# False로 설정할 경우, retrieval 결과들과 Agent의 추론 과정을 보지 않습니다.
pipe.setup() # setup을 통해 crew들을 선언합니다. 이때, 표를 시각화한 그래프를 저장할 output 폴더가 없다면 새로 만듭니다.
```

```python
query = '크래프톤의 핵심사업에 대해 알려줘.'
```

```python
# 보통 QA를 사용하지만, Q만을 따로 하여 retrieval 결과값을 볼 수 있습니다.
result = pipe.Q(query)
```

```python
result = pipe.QA(query, search_type='closed_domain') # search_type을 closed_domain으로 설정할 경우
# 내부 DB를 활용하여 Retrieval을 수행한 뒤, image, table, context 별로 담당 Crew에게 맡겨 답변을 받습니다.
# 이후, 최종 답변 Crew가 세 부분의 답변을 종합해서 합쳐진 결과를 출력합니다.
```

```python
print(result)
```

```python
# search_type = 'open_domain'으로 설정할 경우, 네이버의 뉴스에서 기사를 종합해 답변을 도출합니다.
result = pipe.QA(query, search_type='open_domain')
```
