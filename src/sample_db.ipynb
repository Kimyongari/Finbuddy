```python
from embedding_models.vector_storing import CSVToChromaDB

# Initialize the class
base_dir = "/data/ephemeral/home/datas2"
collection_name = "test_collection"
csv_processor = CSVToChromaDB(base_dir, collection_name)

# Process CSV files
csv_processor.process_all_files()

# Verify the collection
csv_processor.verify_collection()

# Setup retrieval
retriever = csv_processor.setup_retrieval()

# Query retrieval
query_text = "이 회사의 투자 현황은?"
results = csv_processor.query(query_text, retriever)

# Print the retrieved documents
for idx, doc in enumerate(results, start=1):
    print(f"Document {idx}:\n{doc.page_content}\n")
```
